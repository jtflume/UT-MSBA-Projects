```python
import spacy
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
nlp=spacy.load('en_core_web_sm')
```

```python
features = pd.read_csv('all_reviews.csv')
profeatures = features[['Unnamed: 0','company', 'pros']]
confeatures = features[['Unnamed: 0','company', 'cons']]
profeatures
```

```python
diversitypro = ('diveristy', 'inclusive')
environmentpro = ('office', 'perks', 'technoolgy', 'food', 'tech','environment')
culturepro = ('company', 'people', 'culture', 'team','fun','care','friendly','colleagues','supportive','support','remote','values','cares','missoin','collaborative')
worklifepro = ('life','balance','flexible','hours','flexibility','schedule','pto')
paypro = ('pay', 'salary','compensation','stock','benefits','bonuses','options')
careergrowthpro = ('opportunity','growth','opportunity','career','grow')
leadershippro = ('leadership','management','training','managers')
personaldevpro = ('learning','projects','experience','learn','challenging','resources','development')
```

```python
profeatures['token']=profeatures['pros'].map(lambda x:nlp(x))
profeatures.to_pickle('profeatures_token.pkl')
```

```python
def pros(comment):
    new_comment = []
    for word in comment:
        if word.is_punct:
            continue
        elif word.is_stop:
            continue
        elif word.text.lower() in diversitypro:
            word='diversity'
            new_comment.append(word)
        elif word.text.lower() in environmentpro:
            word='environment'
            new_comment.append(word)
        elif word.text.lower() in culturepro:
            word='culture'
            new_comment.append(word)
        elif word.text.lower() in worklifepro:
            word='worklife'
            new_comment.append(word)
        elif word.text.lower() in paypro:
            word='pay'
            new_comment.append(word)
        elif word.text.lower() in careergrowthpro:
            word='growth'
            new_comment.append(word)
        elif word.text.lower() in leadershippro:
            word='leadership'
            new_comment.append(word)
        elif word.text.lower() in personaldevpro:
            word='personaldevelopment'
            new_comment.append(word)
        else: 
            word = word.text.lower()
            new_comment.append(word)
        
    return(' '.join(new_comment))

profeatures['conpros']=profeatures['token'].map(pros)

#profeatures['conpros']=profeatures['conpros'].map(lambda x:nlp(x))
```

```python
attributes = []


for row in range(len(profeatures)):
    text=profeatures['pros'].iloc[row]
    attributelist = list()
    for word in text:
        if word.is_punct:
            continue
        if word.text in featuresnames:
            attributelist.append(word.text)
    attributes.append(attributelist)
attributes
```

```python
def count_firm_atb_pair(firm, attribute):
    count = 0
    for i in range(len(profeatures)):
        if ((firm == profeatures['company'].iloc[i]) and (attribute in profeatures['attributes'].iloc[i])):
            count += 1
    return count
```

```python
def count_firm(firm):
    count = 0
    for i in range(len(features)):
        if (firm == features['company'].iloc[i]):
            count += 1
    return count
```

```python
def count_attribute(attribute):
    count = 0
    for i in range(len(profeatures)):
        if (attribute in profeatures['attributes'].iloc[i]):
            count += 1
    return count
```

```python
def get_lift_atb (firm, attribute):
    numerator = len(features) * count_firm_atb_pair(firm, attribute)
    denominator = count_firm(firm) * count_attribute(attribute)
    if denominator == 0:
        print(firm,attribute)
    return numerator / denominator 
```

```python
atb_list = []
firm_pairs = []
for i in range(len(firmname)):
    firm_a = firmname[i]
    i_list = []
    for j in range(len(featuresnames)):
        attribute = featuresnames[j]
        lift = get_lift_atb(firm_a, attribute)
        i_list.append(lift)
atb_list.append(i_list)
```
